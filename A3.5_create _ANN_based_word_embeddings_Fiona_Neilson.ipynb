## Word embeddings

Input = **df3**

```python
import io
import os
import re
import shutil
import string
from numpy import array
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import TextVectorization
```

```python
# create list of comments
docs = df3['comment'].str.split(',').apply(lambda x: ', '.join([i.strip() for i in x])).tolist()
```

```python
docs
```

```python
# define class labels
labels = df3['stereotype'].to_numpy()
print(labels)
```

```python
unique_words = set(" ".join(docs).split())
vocab_size = len(unique_words)
```

```python
vocab_size
```

```python
# integer encode the documents
encoded_docs = [one_hot(d, vocab_size) for d in docs]
print(encoded_docs)
```

```python
# pad documents to a max length of 20 words
# 19 is the av length of comments
max_length = 20
padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
print(padded_docs)
```

```python
# Create a custom standardization function to strip HTML break tags '<br />'.
def custom_standardization(input_data):
  lowercase = tf.strings.lower(input_data)
  stripped_html = tf.strings.regex_replace(lowercase, '<br />', ' ')
  return tf.strings.regex_replace(stripped_html,
                                  '[%s]' % re.escape(string.punctuation), '')
```

### Create vector to word mapping

```python
# Vocabulary size and number of words in a sequence.
sequence_length = 20

# Use the text vectorization layer to normalize, split, and map strings to
# integers. Note that the layer uses the custom standardization defined above.
# Set maximum_sequence length as all samples are not of the same length.
vectorize_layer = TextVectorization(
    standardize=custom_standardization,
    max_tokens=vocab_size,
    output_mode='int',
    output_sequence_length=sequence_length)

# Make a text-only dataset (no labels) and call adapt to build the vocabulary.
text_ds = docs #train_ds.map(lambda x, y: x)
vectorize_layer.adapt(text_ds)
print(vectorize_layer.get_vocabulary())
```

### Define the model 

Tensorflow is the Google ML framework used to build the Neural Network.  'Sigmoid' is used as the activation function to give probability which gives a result from 0 to 1 and is used for this binary classification task.  

```python
# define the model via add() method
# vocab size 3287
model = Sequential()
model.add(Embedding(vocab_size, 20, name="embedding", input_length=max_length))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid')) # output later
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# summarize the model
print(model.summary())
```

```python
# fit the model
model.fit(padded_docs, labels, epochs=70, verbose=1)
```

```python
# evaluate the model
loss, accuracy = model.evaluate(padded_docs, labels, verbose=0)
print('Accuracy: %f' % (accuracy*100))
```

```python
weights = model.get_layer('embedding').get_weights()[0]
vocab = vectorize_layer.get_vocabulary()
```

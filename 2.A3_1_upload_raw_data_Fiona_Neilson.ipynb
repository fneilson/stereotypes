**IMPORTANT** DO NOT RE-RUN THIS FUNCTION AS IT WILL CHANGE THE ENCODINGS

```python
# create a function to detect language and add to the dataframe
def det(x):
    try:
        lang = detect(x)
    except:
        lang = 'Other'
    return lang

yt['lang'] = yt['comment'].apply(det)
```
